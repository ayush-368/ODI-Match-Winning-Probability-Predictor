```python
import numpy as np
import pandas as pd
import os
from tqdm import tqdm
from yaml import safe_load
```

```python
filenames = []
for file in os.listdir('D:\ML\ODI\data'):
    filenames.append(os.path.join('D:\ML\ODI\data',file))
```

```python
filenames[:5]
```

```python
df = []
counter = 1
for file in tqdm(filenames):
    with open(file,'r') as f:
        temp=pd.json_normalize(safe_load(f))
        temp['id'] = counter
        df.append(temp)
        counter+=1
```

```python
final_df = pd.concat(df,ignore_index=True)
```
