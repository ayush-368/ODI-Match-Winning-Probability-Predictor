```python
del_df = []

for index,row in df.iterrows():
    
    inning_no = []
    batting_team = []
    ball_no = []
    batsman = []
    bowler = []
    non_striker = []
    runs = []
    extra = []
    wicket = []
    city = []
    winner = []
    teams = []
    umpire = []
    venue = []
    match_id = []
    year = []
    
    over_no = 0
    ecnt=0
    
    
    
    
    for i in range (len(row['innings'])):
        for inning in (row['innings'][i].keys()):
            for ball in (row['innings'][i][inning]['deliveries']):
                for key in ball.keys():


                    inning_no.append(inning) #inning no
                    batting_team.append(row['innings'][i][inning]['team']) #batting team
                    batsman.append(ball[key]['batsman']) #batsman
                    bowler.append(ball[key]['bowler'])#bowler
                    non_striker.append(ball[key]['non_striker']) #non striker
                    runs.append(ball[key]['runs']['total']) #total runs in that ball
                    city.append(row['info.city']) #in which city this match is played
                    winner.append(row['info.outcome.winner']) #winner
                    teams.append(row['info.teams']) #teams
                    umpire.append(row['info.umpires'])
                    venue.append(row['info.venue'])
                    match_id.append(row['id'])
                    year.append(row['info.dates'])
                    
                    
                    try:
                        wicket.append(ball[key]['wicket']['player_out'])
                    except:
                        wicket.append('0')
                        
                    try:
                        extra.append(list(ball[key]['extras'].keys())[0])

                    except:
                        extra.append('0')
        
        
        
                    #ball balancing
                    if (key >= over_no and key < over_no+1):
                        over_no=over_no  
                    else:
                        over_no = int(str(key).split('.')[0])
                        ecnt = 0   
                    try:
                        if((list(ball[key]['extras'].keys())[0]=='noballs') or (list(ball[key]['extras'].keys())[0]=='wides')):
                            ecnt+=1
                    except:
                        ecnt=ecnt
                    key = key - (ecnt*0.1)
                    
                    ball_no.append(key) #ball no
                    
    temp_df = pd.DataFrame({
        'match_id':match_id,
        'inning_no':inning_no,
        'batting_team':batting_team,
        'ball_no':ball_no,
        'batsman':batsman,
        'bowler':bowler,
        'non_striker':non_striker,
        'runs':runs,
        'extra':extra,
        'wicket':wicket,
        'city':city,
        'winner':winner,
        'teams':teams,
        'umpire':umpire,
        'venue':venue,
        'date':year
    })
        
    del_df.append(temp_df)
```

```python
def resolve_infinity_val_in_crr(df):
    if(df['balls_thrown']==0):
        return df['score']
    else:
        return df['crr']
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test = train_test_split(final_df,test_size=0.2,random_state=42)
```

```python
y_train = train['result']
x_train = train.drop(columns=['result'])

y_test = test['result']
x_test = test.drop(columns=['result'])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.gaussian_process import GaussianProcessClassifier

from sklearn.metrics import r2_score,mean_absolute_error,accuracy_score
```

```python
step1 = ColumnTransformer(transformers=[
    ('col_tr',OneHotEncoder(sparse=False,drop='first'),[0,1,2])
],remainder='passthrough')

step2 = StandardScaler()

step3 = RandomForestClassifier()
```

```python
pipe = Pipeline(steps=[
    ('step1', step1),
    ('step2',step2),
    ('step3',step3)
])
```

```python
pipe.fit(x_train,y_train)
```

```python
predictions = pipe.predict(x_test)
```

```python
print('r2 score : ',r2_score(y_test,predictions))
print('mean absolute error : ',mean_absolute_error(y_test,predictions))
print('accuracy score : ',accuracy_score(y_test,predictions))
```
